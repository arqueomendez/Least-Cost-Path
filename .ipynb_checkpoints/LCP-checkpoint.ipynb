```python
import numpy as np
import rasterio
import math
import os
from tqdm import tqdm
import fiona
from shapely.geometry import Point, LineString, mapping
from numba import njit
from datetime import datetime
from skimage.draw import disk
from rasterio.enums import Resampling

# --- 1. CONFIGURACIÓN DEL PROYECTO Y PARÁMETROS ---
BASE_PATH = R"C:\Users\User\Desktop\SHanna"
POINTS_SUBFOLDER = os.path.join(BASE_PATH, 'puntos_por_100km')
COST_RASTER_PATH = os.path.join(BASE_PATH, 'final_total_cost.tif')

# --- PARÁMETROS DE ENTRADA ---
START_SHAPEFILE_PATH = os.path.join(POINTS_SUBFOLDER, 'rand_point_0.shp')
END_SHAPEFILE_PATH = os.path.join(POINTS_SUBFOLDER, 'rand_point_2.shp')

# --- PARÁMETROS DE LA BÚSQUEDA JERÁRQUICA ---
DOWNSAMPLING_FACTOR = 32
CORRIDOR_BUFFER_PIXELS = 150
HEURISTIC_WEIGHT = 1.0

# --- 2. FUNCIONES AUXILIARES ---

def get_coords_from_shapefile(shapefile_path):
    try:
        with fiona.open(shapefile_path, 'r') as c:
            feat = next(iter(c)); geom = feat['geometry']
            return geom['coordinates'], c.crs
    except Exception as e:
        print(f"Error leyendo {shapefile_path}: {e}"); return None, None

def world_to_pixel(transform, x, y):
    col, row = ~transform * (x, y); return int(row), int(col)

# --- 3. ALGORITMO A* Y HELPERS (COMPILADOS CON NUMBA) ---

@njit
def heuristic_numba(r1, c1, r2, c2, dx, dy):
    return math.sqrt((r2 - r1)**2 + (c2 - c1)**2) * math.sqrt(dx * dy)

@njit
def a_star_numba_compiled(cost_array, nodata_value, start_pixel, end_pixel, dx, dy, weight, search_mask):
    height, width = cost_array.shape
    g_cost = np.full(cost_array.shape, np.inf, dtype=np.float64)
    came_from = np.full(cost_array.shape, -1, dtype=np.int16)
    open_set = np.zeros((1, 4), dtype=np.float64)
    h_initial = heuristic_numba(start_pixel[0], start_pixel[1], end_pixel[0], end_pixel[1], dx, dy)
    open_set[0] = [h_initial * weight, 0.0, start_pixel[0], start_pixel[1]]
    g_cost[start_pixel] = 0
    path_found = False
    
    while open_set.shape[0] > 0:
        min_idx = np.argmin(open_set[:, 0])
        f, g, r, c = open_set[min_idx]; current_pos = (int(r), int(c))
        if open_set.shape[0] == 1: open_set = np.zeros((0, 4), dtype=np.float64)
        else: open_set = np.vstack((open_set[:min_idx], open_set[min_idx + 1:]))
        
        if current_pos == end_pixel: path_found = True; break
        if g > g_cost[current_pos]: continue
            
        cost_current = cost_array[current_pos]
        
        for dr in range(-1, 2):
            for dc in range(-1, 2):
                if dr == 0 and dc == 0: continue
                neighbor_pos = (current_pos[0] + dr, current_pos[1] + dc)
                if not (0 <= neighbor_pos[0] < height and 0 <= neighbor_pos[1] < width): continue
                if not search_mask[neighbor_pos]: continue
                cost_neighbor = cost_array[neighbor_pos]
                if cost_neighbor == nodata_value: continue
                dist_m = math.sqrt((dr * dy)**2 + (dc * dx)**2)
                avg_cost = (cost_current + cost_neighbor) / 2.0
                tentative_g_cost = g + (avg_cost * dist_m)
                
                if tentative_g_cost < g_cost[neighbor_pos]:
                    direction = (dr+1)*3 + (dc+1)
                    came_from[neighbor_pos] = direction; g_cost[neighbor_pos] = tentative_g_cost
                    h = heuristic_numba(neighbor_pos[0], neighbor_pos[1], end_pixel[0], end_pixel[1], dx, dy)
                    new_f_cost = tentative_g_cost + (h * weight)
                    new_entry = np.array([[new_f_cost, tentative_g_cost, neighbor_pos[0], neighbor_pos[1]]], dtype=np.float64)
                    if open_set.shape[0] == 0: open_set = new_entry
                    else: open_set = np.vstack((open_set, new_entry))
    
    return path_found, came_from, g_cost

@njit
def reconstruct_path_pixels_numba(came_from_array, start_pixel, end_pixel):
    path = np.zeros((came_from_array.size, 2), dtype=np.int32)
    current_pos_r, current_pos_c = end_pixel
    count = 0
    limit = came_from_array.size
    
    while (current_pos_r, current_pos_c) != start_pixel and count < limit:
        path[count] = np.array([current_pos_r, current_pos_c], dtype=np.int32)
        direction = came_from_array[current_pos_r, current_pos_c]
        if direction == -1: return None
        dc = (direction % 3) - 1; dr = (direction // 3) - 1
        current_pos_r -= dr; current_pos_c -= dc
        count += 1
        
    path[count] = np.array([start_pixel[0], start_pixel[1]], dtype=np.int32)
    return path[:count+1][::-1]

# --- 4. FUNCIONES PARA BÚSQUEDA JERÁRQUICA ---

def create_low_res_data(src, factor):
    low_res_shape = (src.height // factor, src.width // factor)
    low_res_data = src.read(1, out_shape=low_res_shape, resampling=Resampling.average)
    low_res_transform = src.transform * src.transform.scale(factor, factor)
    low_res_dx = src.res[0] * factor
    low_res_dy = src.res[1] * factor
    return low_res_data, low_res_transform, low_res_dx, low_res_dy

def create_search_corridor(path_low_res, high_res_shape, factor, buffer_pixels):
    corridor_mask = np.zeros(high_res_shape, dtype=bool)
    if path_low_res is None: return corridor_mask
    for r_low, c_low in path_low_res:
        r_high = int(r_low * factor + factor / 2)
        c_high = int(c_low * factor + factor / 2)
        rr, cc = disk((r_high, c_high), buffer_pixels, shape=high_res_shape)
        corridor_mask[rr, cc] = True
    return corridor_mask

def save_path_to_shapefile(pixel_path, transform, crs, output_path):
    world_coords = [transform * (p[1] + 0.5, p[0] + 0.5) for p in pixel_path]
    schema = {'geometry': 'LineString', 'properties': {'id': 'str'}}
    with fiona.open(output_path, 'w', 'ESRI Shapefile', schema, fiona.crs.from_string(str(crs))) as c:
        c.write({'geometry': mapping(LineString(world_coords)), 'properties': {'id': os.path.basename(output_path)}})

# --- 5. EJECUCIÓN DEL ANÁLISIS JERÁRQUICO PARA UNA RUTA ---
if __name__ == '__main__':
    try:
        # --- MODIFICACIÓN: Crear directorio de salida dinámico ---
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        OUTPUT_DIR = os.path.join(BASE_PATH, f"ruta_test_{timestamp}")
        os.makedirs(OUTPUT_DIR, exist_ok=True)
        print(f"Los resultados de esta prueba se guardarán en: {OUTPUT_DIR}")

        # --- MODIFICACIÓN: Actualizar rutas de salida ---
        OUTPUT_SHAPEFILE_PHASE1_PATH = os.path.join(OUTPUT_DIR, 'ruta_jerarquica_FASE1_TEST.shp')
        OUTPUT_SHAPEFILE_PATH = os.path.join(OUTPUT_DIR, 'ruta_jerarquica_FINAL_TEST.shp')

        start_coords, start_crs = get_coords_from_shapefile(START_SHAPEFILE_PATH)
        end_coords, end_crs = get_coords_from_shapefile(END_SHAPEFILE_PATH)

        if start_coords and end_coords:
            with rasterio.open(COST_RASTER_PATH) as src:
                if not (src.crs == start_crs and src.crs == end_crs):
                    print("¡ERROR CRÍTICO: Los CRS no coinciden!")
                else:
                    print("\n--- INICIANDO PRUEBA DE RUTA JERÁRQUICA ---")
                    
                    print("Cargando superficie de coste de alta resolución en RAM...")
                    cost_data_high_res = src.read(1)
                    print(f"Creando superficie de coste de baja resolución (factor {DOWNSAMPLING_FACTOR}x)...")
                    cost_data_low_res, trans_low, dx_low, dy_low = create_low_res_data(src, DOWNSAMPLING_FACTOR)
                    
                    start_pixel_hr = world_to_pixel(src.transform, start_coords[0], start_coords[1])
                    end_pixel_hr = world_to_pixel(src.transform, end_coords[0], end_coords[1])
                    start_pixel_lr = (start_pixel_hr[0] // DOWNSAMPLING_FACTOR, start_pixel_hr[1] // DOWNSAMPLING_FACTOR)
                    end_pixel_lr = (end_pixel_hr[0] // DOWNSAMPLING_FACTOR, end_pixel_hr[1] // DOWNSAMPLING_FACTOR)
                    
                    print(f"\nFASE 1: Ejecutando A* en el ráster de baja resolución ({cost_data_low_res.shape[0]}x{cost_data_low_res.shape[1]} píxeles)...")
                    search_mask_low_res = np.ones(cost_data_low_res.shape, dtype=bool)
                    path_found_lr, came_from_lr, _ = a_star_numba_compiled(
                        cost_data_low_res, src.nodata, start_pixel_lr, end_pixel_lr, dx_low, abs(dy_low), HEURISTIC_WEIGHT, search_mask_low_res
                    )
                    
                    if not path_found_lr:
                        print("ERROR FATAL: No se pudo encontrar una ruta ni siquiera en baja resolución.")
                    else:
                        print("FASE 1 Completada: Ruta estratégica encontrada.")
                        path_pixels_lr = reconstruct_path_pixels_numba(came_from_lr, start_pixel_lr, end_pixel_lr)
                        
                        print("Guardando el shapefile de la ruta de Fase 1 (baja resolución)...")
                        save_path_to_shapefile(path_pixels_lr, trans_low, src.crs, OUTPUT_SHAPEFILE_PHASE1_PATH)
                        print(f"  - Ruta de Fase 1 guardada en: {os.path.basename(OUTPUT_SHAPEFILE_PHASE1_PATH)}")

                        print(f"\nCreando corredor de búsqueda (ancho: {CORRIDOR_BUFFER_PIXELS} píxeles)...")
                        search_mask_hr = create_search_corridor(path_pixels_lr, cost_data_high_res.shape, DOWNSAMPLING_FACTOR, CORRIDOR_BUFFER_PIXELS)
                        
                        print(f"\nFASE 2: Ejecutando A* en el ráster de alta resolución (dentro del corredor)...")
                        path_found_hr, came_from_hr, g_cost_hr = a_star_numba_compiled(
                            cost_data_high_res, src.nodata, start_pixel_hr, end_pixel_hr, src.res[0], abs(src.res[1]), HEURISTIC_WEIGHT, search_mask_hr
                        )
                        
                        if path_found_hr:
                            print("FASE 2 Completada: Ruta detallada encontrada.")
                            print("\nReconstruyendo y guardando el shapefile final...")
                            final_cost = g_cost_hr[end_pixel_hr]
                            path_pixels_hr = reconstruct_path_pixels_numba(came_from_hr, start_pixel_hr, end_pixel_hr)
                            save_path_to_shapefile(path_pixels_hr, src.transform, src.crs, OUTPUT_SHAPEFILE_PATH)
                            print(f"\n¡PRUEBA COMPLETADA CON ÉXITO!")
                            print(f"  - Ruta final guardada en: {os.path.basename(OUTPUT_SHAPEFILE_PATH)}")
                            print(f"  - Coste total de la ruta: {final_cost:.2f}")
                        else:
                            print("\nERROR FATAL: No se pudo encontrar una ruta en alta resolución dentro del corredor.")
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")
```

```python
import numpy as np
import rasterio
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib_scalebar.scalebar import ScaleBar
import os

# --- 1. CONFIGURACIÓN ---

BASE_PATH = R"C:\Users\User\Desktop\SHanna"

# --- ¡IMPORTANTE! ACTUALIZA ESTA CARPETA ---
# Apunta al directorio específico con fecha y hora que se creó en la prueba anterior.
# Ejemplo: 'ruta_test_20250905_153000'
TEST_RESULTS_FOLDER = os.path.join(BASE_PATH, 'ruta_test_20250904_160034import numpy as np
import rasterio
import math
import os
from tqdm import tqdm
import fiona
from shapely.geometry import Point, LineString, mapping
from numba import njit
from datetime import datetime
from skimage.draw import disk
from rasterio.enums import Resampling

# --- 1. CONFIGURACIÓN DEL PROYECTO Y PARÁMETROS ---
BASE_PATH = R"C:\Users\LAV\Desktop\SHanna"
POINTS_SUBFOLDER = os.path.join(BASE_PATH, 'puntos_por_100km')
COST_RASTER_PATH = os.path.join(BASE_PATH, 'final_total_cost.tif')

# --- PARÁMETROS DE ENTRADA Y SALIDA ---
# Nombre del Shapefile que contiene TODOS los 11 puntos
ALL_POINTS_SHAPEFILE = os.path.join(POINTS_SUBFOLDER, 'puntos_por_100km.shp') # <--- CONFIRMA ESTE NOMBRE

# ID del punto que será el ORIGEN de todas las rutas
ORIGIN_POINT_ID = 0 # <--- PUNTO DE INICIO FIJO

# Nombre del campo en el shapefile que contiene los IDs de los puntos
ID_FIELD_NAME = 'ID' # <--- CONFIRMA EL NOMBRE DE ESTE CAMPO

# --- PARÁMETROS DE LA BÚSQUEDA JERÁRQUICA ---
DOWNSAMPLING_FACTOR = 32
CORRIDOR_BUFFER_PIXELS = 150
HEURISTIC_WEIGHT = 1.0

# --- 2. FUNCIONES AUXILIARES ---

def load_all_points(shapefile_path, id_field):
    """Carga todos los puntos y sus IDs desde un shapefile a un diccionario."""
    points = {}
    crs = None
    try:
        with fiona.open(shapefile_path, 'r') as collection:
            crs = collection.crs
            if len(collection) == 0:
                print(f"Error: El shapefile '{os.path.basename(shapefile_path)}' está vacío.")
                return None, None
            for feature in collection:
                # Asegurarse de que el ID es un tipo de dato consistente (entero)
                point_id = int(feature['properties'][id_field])
                coords = feature['geometry']['coordinates']
                points[point_id] = coords
        return points, crs
    except Exception as e:
        print(f"Error cargando los puntos desde {shapefile_path}: {e}")
        return None, None

def world_to_pixel(transform, x, y):
    col, row = ~transform * (x, y); return int(row), int(col)

# --- 3. ALGORITMO A* Y HELPERS (COMPILADOS CON NUMBA) ---

@njit
def heuristic_numba(r1, c1, r2, c2, dx, dy):
    return math.sqrt((r2 - r1)**2 + (c2 - c1)**2) * math.sqrt(dx * dy)

@njit
def a_star_numba_compiled(cost_array, nodata_value, start_pixel, end_pixel, dx, dy, weight, search_mask):
    height, width = cost_array.shape
    g_cost = np.full(cost_array.shape, np.inf, dtype=np.float64)
    came_from = np.full(cost_array.shape, -1, dtype=np.int16)
    open_set = np.zeros((1, 4), dtype=np.float64)
    h_initial = heuristic_numba(start_pixel[0], start_pixel[1], end_pixel[0], end_pixel[1], dx, dy)
    open_set[0] = [h_initial * weight, 0.0, start_pixel[0], start_pixel[1]]
    g_cost[start_pixel] = 0
    path_found = False
    
    while open_set.shape[0] > 0:
        min_idx = np.argmin(open_set[:, 0])
        f, g, r, c = open_set[min_idx]; current_pos = (int(r), int(c))
        if open_set.shape[0] == 1: open_set = np.zeros((0, 4), dtype=np.float64)
        else: open_set = np.vstack((open_set[:min_idx], open_set[min_idx + 1:]))
        
        if current_pos == end_pixel: path_found = True; break
        if g > g_cost[current_pos]: continue
            
        cost_current = cost_array[current_pos]
        
        for dr in range(-1, 2):
            for dc in range(-1, 2):
                if dr == 0 and dc == 0: continue
                neighbor_pos = (current_pos[0] + dr, current_pos[1] + dc)
                if not (0 <= neighbor_pos[0] < height and 0 <= neighbor_pos[1] < width): continue
                if not search_mask[neighbor_pos]: continue
                cost_neighbor = cost_array[neighbor_pos]
                if cost_neighbor == nodata_value: continue
                dist_m = math.sqrt((dr * dy)**2 + (dc * dx)**2)
                avg_cost = (cost_current + cost_neighbor) / 2.0
                tentative_g_cost = g + (avg_cost * dist_m)
                
                if tentative_g_cost < g_cost[neighbor_pos]:
                    direction = (dr+1)*3 + (dc+1)
                    came_from[neighbor_pos] = direction; g_cost[neighbor_pos] = tentative_g_cost
                    h = heuristic_numba(neighbor_pos[0], neighbor_pos[1], end_pixel[0], end_pixel[1], dx, dy)
                    new_f_cost = tentative_g_cost + (h * weight)
                    new_entry = np.array([[new_f_cost, tentative_g_cost, neighbor_pos[0], neighbor_pos[1]]], dtype=np.float64)
                    if open_set.shape[0] == 0: open_set = new_entry
                    else: open_set = np.vstack((open_set, new_entry))
    
    return path_found, came_from, g_cost

@njit
def reconstruct_path_pixels_numba(came_from_array, start_pixel, end_pixel):
    path = np.zeros((came_from_array.size, 2), dtype=np.int32)
    current_pos_r, current_pos_c = end_pixel
    count = 0
    limit = came_from_array.size
    
    while (current_pos_r, current_pos_c) != start_pixel and count < limit:
        path[count] = np.array([current_pos_r, current_pos_c], dtype=np.int32)
        direction = came_from_array[current_pos_r, current_pos_c]
        if direction == -1: return None
        dc = (direction % 3) - 1; dr = (direction // 3) - 1
        current_pos_r -= dr; current_pos_c -= dc
        count += 1
        
    path[count] = np.array([start_pixel[0], start_pixel[1]], dtype=np.int32)
    return path[:count+1][::-1]

# --- 4. FUNCIONES PARA BÚSQUEDA JERÁRQUICA ---

def create_low_res_data(src, factor):
    low_res_shape = (src.height // factor, src.width // factor)
    low_res_data = src.read(1, out_shape=low_res_shape, resampling=Resampling.average)
    low_res_transform = src.transform * src.transform.scale(factor, factor)
    low_res_dx = src.res[0] * factor
    low_res_dy = src.res[1] * factor
    return low_res_data, low_res_transform, low_res_dx, low_res_dy

def create_search_corridor(path_low_res, high_res_shape, factor, buffer_pixels):
    corridor_mask = np.zeros(high_res_shape, dtype=bool)
    if path_low_res is None: return corridor_mask
    for r_low, c_low in path_low_res:
        r_high = int(r_low * factor + factor / 2)
        c_high = int(c_low * factor + factor / 2)
        rr, cc = disk((r_high, c_high), buffer_pixels, shape=high_res_shape)
        corridor_mask[rr, cc] = True
    return corridor_mask

def save_path_to_shapefile(pixel_path, transform, crs, output_path):
    world_coords = [transform * (p[1] + 0.5, p[0] + 0.5) for p in pixel_path]
    schema = {'geometry': 'LineString', 'properties': {'id': 'str'}}
    with fiona.open(output_path, 'w', 'ESRI Shapefile', schema, fiona.crs.from_string(str(crs))) as c:
        c.write({'geometry': mapping(LineString(world_coords)), 'properties': {'id': os.path.basename(output_path)}})

# --- 5. EJECUCIÓN DEL ANÁLISIS DE UNO A TODOS ---
if __name__ == '__main__':
    try:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        OUTPUT_DIR = os.path.join(BASE_PATH, f"rutas_desde_punto_{ORIGIN_POINT_ID}_{timestamp}")
        os.makedirs(OUTPUT_DIR, exist_ok=True)
        print(f"Los resultados se guardarán en: {OUTPUT_DIR}")
        
        all_points, points_crs = load_all_points(ALL_POINTS_SHAPEFILE, ID_FIELD_NAME)

        if all_points:
            with rasterio.open(COST_RASTER_PATH) as src:
                if not (src.crs == points_crs):
                    print("¡ERROR CRÍTICO: El CRS de los puntos y el del ráster no coinciden!")
                else:
                    print("Cargando superficie de coste de alta resolución en RAM...")
                    cost_data_high_res = src.read(1)
                    print(f"Creando superficie de coste de baja resolución (factor {DOWNSAMPLING_FACTOR}x)...")
                    cost_data_low_res, trans_low, dx_low, dy_low = create_low_res_data(src, DOWNSAMPLING_FACTOR)
                    
                    origin_coords = all_points.get(ORIGIN_POINT_ID)
                    if not origin_coords:
                        print(f"Error: No se encontró el punto de origen con ID={ORIGIN_POINT_ID} en el shapefile.")
                    else:
                        start_pixel_hr = world_to_pixel(src.transform, origin_coords[0], origin_coords[1])
                        print(f"\nOrigen Fijo: Punto ID {ORIGIN_POINT_ID} -> Píxel {start_pixel_hr}")
                        
                        for dest_id, dest_coords in tqdm(all_points.items(), desc="Calculando rutas"):
                            if dest_id == ORIGIN_POINT_ID:
                                continue
                            
                            end_pixel_hr = world_to_pixel(src.transform, dest_coords[0], dest_coords[1])
                            
                            start_pixel_lr = (start_pixel_hr[0] // DOWNSAMPLING_FACTOR, start_pixel_hr[1] // DOWNSAMPLING_FACTOR)
                            end_pixel_lr = (end_pixel_hr[0] // DOWNSAMPLING_FACTOR, end_pixel_hr[1] // DOWNSAMPLING_FACTOR)
                            
                            search_mask_low_res = np.ones(cost_data_low_res.shape, dtype=bool)
                            path_found_lr, came_from_lr, _ = a_star_numba_compiled(
                                cost_data_low_res, src.nodata, start_pixel_lr, end_pixel_lr, dx_low, abs(dy_low), HEURISTIC_WEIGHT, search_mask_low_res
                            )
                            
                            if not path_found_lr:
                                print(f"  ADVERTENCIA: No se encontró ruta de bajo nivel para {ORIGIN_POINT_ID}->{dest_id}")
                                continue
                            
                            path_pixels_lr = reconstruct_path_pixels_numba(came_from_lr, start_pixel_lr, end_pixel_lr)
                            search_mask_hr = create_search_corridor(path_pixels_lr, cost_data_high_res.shape, DOWNSAMPLING_FACTOR, CORRIDOR_BUFFER_PIXELS)
                            
                            path_found_hr, came_from_hr, g_cost_hr = a_star_numba_compiled(
                                cost_data_high_res, src.nodata, start_pixel_hr, end_pixel_hr, src.res[0], abs(src.res[1]), HEURISTIC_WEIGHT, search_mask_hr
                            )
                            
                            if path_found_hr:
                                path_pixels_hr = reconstruct_path_pixels_numba(came_from_hr, start_pixel_hr, end_pixel_hr)
                                
                                # Guardar ruta de Fase 1
                                output_shp_p1 = os.path.join(OUTPUT_DIR, f"ruta_fase1_desde_{ORIGIN_POINT_ID}_a_{dest_id}.shp")
                                save_path_to_shapefile(path_pixels_lr, trans_low, src.crs, output_shp_p1)
                                
                                # Guardar ruta Final
                                output_shp_final = os.path.join(OUTPUT_DIR, f"ruta_final_desde_{ORIGIN_POINT_ID}_a_{dest_id}.shp")
                                save_path_to_shapefile(path_pixels_hr, src.transform, src.crs, output_shp_final)
                            else:
                                print(f"  ADVERTENCIA: No se encontró ruta de alto nivel para {ORIGIN_POINT_ID}->{dest_id}")

                        print("\n¡Proceso de Uno a Todos completado!")

    except Exception as e:
        print(f"Ocurrió un error inesperado en el proceso principal: {e}")')

# Rutas a los archivos de entrada
COST_RASTER_PATH = os.path.join(BASE_PATH, 'final_total_cost.tif')
PHASE_1_ROUTE_PATH = os.path.join(TEST_RESULTS_FOLDER, 'ruta_jerarquica_FASE1_TEST.shp')
FINAL_ROUTE_PATH = os.path.join(TEST_RESULTS_FOLDER, 'ruta_jerarquica_FINAL_TEST.shp')

# Margen para la visualización
BUFFER_KM = 5
BUFFER_METERS = BUFFER_KM * 1000

# --- 2. FUNCIONES DE VISUALIZACIÓN ---

def plot_results(cost_raster_path, phase1_shp_path, final_shp_path):
    """
    Carga el ráster de coste y las dos rutas, y crea un mapa comparativo
    con una extensión ajustada automáticamente.
    """
    # --- Validar que los archivos existen ---
    for path in [cost_raster_path, phase1_shp_path, final_shp_path]:
        if not os.path.exists(path):
            print(f"Error: No se encontró el archivo de entrada en: {path}")
            if "AQUÍ_VA_LA_CARPETA" in path:
                print("Por favor, actualiza la variable 'TEST_RESULTS_FOLDER' con el nombre correcto de la carpeta.")
            return

    # --- Cargar las rutas y calcular la extensión combinada ---
    print("Cargando las rutas vectoriales...")
    route_p1 = gpd.read_file(phase1_shp_path)
    route_final = gpd.read_file(final_shp_path)
    
    # Combinar para obtener los límites totales
    combined_routes = gpd.pd.concat([route_p1, route_final])
    minx, miny, maxx, maxy = combined_routes.total_bounds
    
    # --- Aplicar el buffer de 5 km a la extensión ---
    extent_buffered = [
        minx - BUFFER_METERS,
        maxx + BUFFER_METERS,
        miny - BUFFER_METERS,
        maxy + BUFFER_METERS
    ]
    
    # --- Leer solo la porción necesaria del ráster de coste ---
    print("Leyendo la porción relevante del ráster de coste...")
    with rasterio.open(cost_raster_path) as src:
        # Convertir las coordenadas del mundo a la ventana de píxeles
        window = src.window(extent_buffered[0], extent_buffered[2], extent_buffered[1], extent_buffered[3])
        
        # Obtener la transformación para esta ventana específica
        transform = src.window_transform(window)
        
        # Leer los datos de la ventana
        raster_subset = src.read(1, window=window)

    # --- Crear el Mapa ---
    print("Generando el mapa de visualización...")
    fig, ax = plt.subplots(figsize=(15, 15))

    # 1. Dibujar el ráster de coste como fondo
    # Usamos LogNorm porque los datos de coste están muy sesgados hacia valores bajos
    im = ax.imshow(raster_subset, cmap='viridis', 
                   extent=[extent_buffered[0], extent_buffered[1], extent_buffered[2], extent_buffered[3]],
                   norm=LogNorm())
    fig.colorbar(im, ax=ax, shrink=0.6, label='Coste de Fricción (Escala Logarítmica)')

    # 2. Dibujar las rutas encima
    route_p1.plot(ax=ax, color='cyan', linestyle='--', linewidth=2.5, label='Ruta Fase 1 (Estratégica)')
    route_final.plot(ax=ax, color='magenta', linestyle='-', linewidth=2.5, label='Ruta Final (Detallada)')

    # 3. Añadir elementos cartográficos
    ax.set_title('Comparación de Rutas Jerárquicas (Fase 1 vs. Final)', fontsize=18, pad=20)
    ax.set_xlabel('Coordenada X (Este) [metros]')
    ax.set_ylabel('Coordenada Y (Norte) [metros]')
    ax.legend(loc='upper right')
    ax.grid(True, alpha=0.3)
    
    # Añadir barra de escala (la resolución del píxel es ~29.43m)
    scalebar = ScaleBar(29.43, 'm', length_fraction=0.25, location='lower right')
    ax.add_artist(scalebar)

    # Añadir flecha de norte
    x, y, arrow_len = 0.95, 0.95, 0.08
    ax.annotate('N', xy=(x, y), xytext=(x, y - arrow_len),
                arrowprops=dict(facecolor='black', width=5, headwidth=15),
                ha='center', va='center', fontsize=20,
                xycoords=ax.transAxes)

    plt.tight_layout()
    plt.show()

# --- 3. EJECUCIÓN DEL SCRIPT DE VISUALIZACIÓN ---
if __name__ == '__main__':
    plot_results(COST_RASTER_PATH, PHASE_1_ROUTE_PATH, FINAL_ROUTE_PATH)
```

```python
import numpy as np
import rasterio
import math
import os
from tqdm import tqdm
import fiona
from shapely.geometry import Point, LineString, mapping
from numba import njit
from datetime import datetime
from skimage.draw import disk
from rasterio.enums import Resampling

# --- 1. CONFIGURACIÓN DEL PROYECTO Y PARÁMETROS ---
BASE_PATH = R"C:\Users\User\Desktop\SHanna"
POINTS_SUBFOLDER = os.path.join(BASE_PATH, 'puntos_por_100km')
COST_RASTER_PATH = os.path.join(BASE_PATH, 'final_total_cost.tif')

# --- PARÁMETROS DE ENTRADA Y SALIDA ---
# Nombre del Shapefile que contiene TODOS los 11 puntos
ALL_POINTS_SHAPEFILE = os.path.join(POINTS_SUBFOLDER, 'puntos_por_100km.shp') # <--- CONFIRMA ESTE NOMBRE

# ID del punto que será el ORIGEN de todas las rutas
ORIGIN_POINT_ID = 11 # <--- PUNTO DE INICIO FIJO

# Nombre del campo en el shapefile que contiene los IDs de los puntos
ID_FIELD_NAME = 'rand_point' # <--- CONFIRMA EL NOMBRE DE ESTE CAMPO

# --- PARÁMETROS DE LA BÚSQUEDA JERÁRQUICA ---
DOWNSAMPLING_FACTOR = 32
CORRIDOR_BUFFER_PIXELS = 150
HEURISTIC_WEIGHT = 1.0

# --- 2. FUNCIONES AUXILIARES ---

def load_all_points(shapefile_path, id_field):
    """Carga todos los puntos y sus IDs desde un shapefile a un diccionario."""
    points = {}
    crs = None
    try:
        with fiona.open(shapefile_path, 'r') as collection:
            crs = collection.crs
            if len(collection) == 0:
                print(f"Error: El shapefile '{os.path.basename(shapefile_path)}' está vacío.")
                return None, None
            for feature in collection:
                # Asegurarse de que el ID es un tipo de dato consistente (entero)
                point_id = int(feature['properties'][id_field])
                coords = feature['geometry']['coordinates']
                points[point_id] = coords
        return points, crs
    except Exception as e:
        print(f"Error cargando los puntos desde {shapefile_path}: {e}")
        return None, None

def world_to_pixel(transform, x, y):
    col, row = ~transform * (x, y); return int(row), int(col)

# --- 3. ALGORITMO A* Y HELPERS (COMPILADOS CON NUMBA) ---

@njit
def heuristic_numba(r1, c1, r2, c2, dx, dy):
    return math.sqrt((r2 - r1)**2 + (c2 - c1)**2) * math.sqrt(dx * dy)

@njit
def a_star_numba_compiled(cost_array, nodata_value, start_pixel, end_pixel, dx, dy, weight, search_mask):
    height, width = cost_array.shape
    g_cost = np.full(cost_array.shape, np.inf, dtype=np.float64)
    came_from = np.full(cost_array.shape, -1, dtype=np.int16)
    open_set = np.zeros((1, 4), dtype=np.float64)
    h_initial = heuristic_numba(start_pixel[0], start_pixel[1], end_pixel[0], end_pixel[1], dx, dy)
    open_set[0] = [h_initial * weight, 0.0, start_pixel[0], start_pixel[1]]
    g_cost[start_pixel] = 0
    path_found = False
    
    while open_set.shape[0] > 0:
        min_idx = np.argmin(open_set[:, 0])
        f, g, r, c = open_set[min_idx]; current_pos = (int(r), int(c))
        if open_set.shape[0] == 1: open_set = np.zeros((0, 4), dtype=np.float64)
        else: open_set = np.vstack((open_set[:min_idx], open_set[min_idx + 1:]))
        
        if current_pos == end_pixel: path_found = True; break
        if g > g_cost[current_pos]: continue
            
        cost_current = cost_array[current_pos]
        
        for dr in range(-1, 2):
            for dc in range(-1, 2):
                if dr == 0 and dc == 0: continue
                neighbor_pos = (current_pos[0] + dr, current_pos[1] + dc)
                if not (0 <= neighbor_pos[0] < height and 0 <= neighbor_pos[1] < width): continue
                if not search_mask[neighbor_pos]: continue
                cost_neighbor = cost_array[neighbor_pos]
                if cost_neighbor == nodata_value: continue
                dist_m = math.sqrt((dr * dy)**2 + (dc * dx)**2)
                avg_cost = (cost_current + cost_neighbor) / 2.0
                tentative_g_cost = g + (avg_cost * dist_m)
                
                if tentative_g_cost < g_cost[neighbor_pos]:
                    direction = (dr+1)*3 + (dc+1)
                    came_from[neighbor_pos] = direction; g_cost[neighbor_pos] = tentative_g_cost
                    h = heuristic_numba(neighbor_pos[0], neighbor_pos[1], end_pixel[0], end_pixel[1], dx, dy)
                    new_f_cost = tentative_g_cost + (h * weight)
                    new_entry = np.array([[new_f_cost, tentative_g_cost, neighbor_pos[0], neighbor_pos[1]]], dtype=np.float64)
                    if open_set.shape[0] == 0: open_set = new_entry
                    else: open_set = np.vstack((open_set, new_entry))
    
    return path_found, came_from, g_cost

@njit
def reconstruct_path_pixels_numba(came_from_array, start_pixel, end_pixel):
    path = np.zeros((came_from_array.size, 2), dtype=np.int32)
    current_pos_r, current_pos_c = end_pixel
    count = 0
    limit = came_from_array.size
    
    while (current_pos_r, current_pos_c) != start_pixel and count < limit:
        path[count] = np.array([current_pos_r, current_pos_c], dtype=np.int32)
        direction = came_from_array[current_pos_r, current_pos_c]
        if direction == -1: return None
        dc = (direction % 3) - 1; dr = (direction // 3) - 1
        current_pos_r -= dr; current_pos_c -= dc
        count += 1
        
    path[count] = np.array([start_pixel[0], start_pixel[1]], dtype=np.int32)
    return path[:count+1][::-1]

# --- 4. FUNCIONES PARA BÚSQUEDA JERÁRQUICA ---

def create_low_res_data(src, factor):
    low_res_shape = (src.height // factor, src.width // factor)
    low_res_data = src.read(1, out_shape=low_res_shape, resampling=Resampling.average)
    low_res_transform = src.transform * src.transform.scale(factor, factor)
    low_res_dx = src.res[0] * factor
    low_res_dy = src.res[1] * factor
    return low_res_data, low_res_transform, low_res_dx, low_res_dy

def create_search_corridor(path_low_res, high_res_shape, factor, buffer_pixels):
    corridor_mask = np.zeros(high_res_shape, dtype=bool)
    if path_low_res is None: return corridor_mask
    for r_low, c_low in path_low_res:
        r_high = int(r_low * factor + factor / 2)
        c_high = int(c_low * factor + factor / 2)
        rr, cc = disk((r_high, c_high), buffer_pixels, shape=high_res_shape)
        corridor_mask[rr, cc] = True
    return corridor_mask

def save_path_to_shapefile(pixel_path, transform, crs, output_path):
    world_coords = [transform * (p[1] + 0.5, p[0] + 0.5) for p in pixel_path]
    schema = {'geometry': 'LineString', 'properties': {'id': 'str'}}
    with fiona.open(output_path, 'w', 'ESRI Shapefile', schema, fiona.crs.from_string(str(crs))) as c:
        c.write({'geometry': mapping(LineString(world_coords)), 'properties': {'id': os.path.basename(output_path)}})

# --- 5. EJECUCIÓN DEL ANÁLISIS DE UNO A TODOS ---
if __name__ == '__main__':
    try:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        OUTPUT_DIR = os.path.join(BASE_PATH, f"rutas_desde_punto_{ORIGIN_POINT_ID}_{timestamp}")
        os.makedirs(OUTPUT_DIR, exist_ok=True)
        print(f"Los resultados se guardarán en: {OUTPUT_DIR}")
        
        all_points, points_crs = load_all_points(ALL_POINTS_SHAPEFILE, ID_FIELD_NAME)

        if all_points:
            with rasterio.open(COST_RASTER_PATH) as src:
                if not (src.crs == points_crs):
                    print("¡ERROR CRÍTICO: El CRS de los puntos y el del ráster no coinciden!")
                else:
                    print("Cargando superficie de coste de alta resolución en RAM...")
                    cost_data_high_res = src.read(1)
                    print(f"Creando superficie de coste de baja resolución (factor {DOWNSAMPLING_FACTOR}x)...")
                    cost_data_low_res, trans_low, dx_low, dy_low = create_low_res_data(src, DOWNSAMPLING_FACTOR)
                    
                    origin_coords = all_points.get(ORIGIN_POINT_ID)
                    if not origin_coords:
                        print(f"Error: No se encontró el punto de origen con ID={ORIGIN_POINT_ID} en el shapefile.")
                    else:
                        start_pixel_hr = world_to_pixel(src.transform, origin_coords[0], origin_coords[1])
                        print(f"\nOrigen Fijo: Punto ID {ORIGIN_POINT_ID} -> Píxel {start_pixel_hr}")
                        
                        for dest_id, dest_coords in tqdm(all_points.items(), desc="Calculando rutas"):
                            if dest_id == ORIGIN_POINT_ID:
                                continue
                            
                            end_pixel_hr = world_to_pixel(src.transform, dest_coords[0], dest_coords[1])
                            
                            start_pixel_lr = (start_pixel_hr[0] // DOWNSAMPLING_FACTOR, start_pixel_hr[1] // DOWNSAMPLING_FACTOR)
                            end_pixel_lr = (end_pixel_hr[0] // DOWNSAMPLING_FACTOR, end_pixel_hr[1] // DOWNSAMPLING_FACTOR)
                            
                            search_mask_low_res = np.ones(cost_data_low_res.shape, dtype=bool)
                            path_found_lr, came_from_lr, _ = a_star_numba_compiled(
                                cost_data_low_res, src.nodata, start_pixel_lr, end_pixel_lr, dx_low, abs(dy_low), HEURISTIC_WEIGHT, search_mask_low_res
                            )
                            
                            if not path_found_lr:
                                print(f"  ADVERTENCIA: No se encontró ruta de bajo nivel para {ORIGIN_POINT_ID}->{dest_id}")
                                continue
                            
                            path_pixels_lr = reconstruct_path_pixels_numba(came_from_lr, start_pixel_lr, end_pixel_lr)
                            search_mask_hr = create_search_corridor(path_pixels_lr, cost_data_high_res.shape, DOWNSAMPLING_FACTOR, CORRIDOR_BUFFER_PIXELS)
                            
                            path_found_hr, came_from_hr, g_cost_hr = a_star_numba_compiled(
                                cost_data_high_res, src.nodata, start_pixel_hr, end_pixel_hr, src.res[0], abs(src.res[1]), HEURISTIC_WEIGHT, search_mask_hr
                            )
                            
                            if path_found_hr:
                                path_pixels_hr = reconstruct_path_pixels_numba(came_from_hr, start_pixel_hr, end_pixel_hr)
                                
                                # Guardar ruta de Fase 1
                                output_shp_p1 = os.path.join(OUTPUT_DIR, f"ruta_fase1_desde_{ORIGIN_POINT_ID}_a_{dest_id}.shp")
                                save_path_to_shapefile(path_pixels_lr, trans_low, src.crs, output_shp_p1)
                                
                                # Guardar ruta Final
                                output_shp_final = os.path.join(OUTPUT_DIR, f"ruta_final_desde_{ORIGIN_POINT_ID}_a_{dest_id}.shp")
                                save_path_to_shapefile(path_pixels_hr, src.transform, src.crs, output_shp_final)
                            else:
                                print(f"  ADVERTENCIA: No se encontró ruta de alto nivel para {ORIGIN_POINT_ID}->{dest_id}")

                        print("\n¡Proceso de Uno a Todos completado!")

    except Exception as e:
        print(f"Ocurrió un error inesperado en el proceso principal: {e}")
```
